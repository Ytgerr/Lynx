```python
import feedparser
import csv
import pandas as pd

our_feeds = {
    'Kommersant Economy': 'https://www.kommersant.ru/RSS/economy.xml',
    'Vedomosti Politics': 'https://www.vedomosti.ru/rss/politics',
    'Vedomosti Finance': 'https://www.vedomosti.ru/rss/finance',
    'RBC News': 'https://rssexport.rbc.ru/rbcnews/news/30/full.rss',
    'Interfax': 'https://www.interfax.ru/rss.asp',
    'RIA Novosti': 'https://ria.ru/export/rss2/index.xml',
    'Lenta Politics': 'https://lenta.ru/rss/news/politics/',
    'TASS': 'https://tass.ru/rss/v2.xml',
    'Rossiyskaya Gazeta': 'https://rg.ru/xml/index.xml',
    'Gazeta Ru': 'https://www.gazeta.ru/export/rss/first.xml'
}

f_all_news = 'russian_finance_politics_news.csv'  # Имя файла датасета

# Функции для извлечения данных (с фильтром на description)
def check_url(url_feed):
    return feedparser.parse(url_feed)

def getHeadlines(url_feed):
    headlines = []
    feed = check_url(url_feed)
    for item in feed['items']:
        headlines.append(item.get('title', ''))
    return headlines

def getDescriptions(url_feed):
    descriptions = []
    feed = check_url(url_feed)
    for item in feed['items']:
        desc = item.get('description', '').strip()
        if desc:  # Только если description не пустое
            descriptions.append(desc)
        else:
            descriptions.append(None)  # Placeholder для синхронизации
    return descriptions

def getLinks(url_feed):
    links = []
    feed = check_url(url_feed)
    for item in feed['items']:
        links.append(item.get('link', ''))
    return links

def getDates(url_feed):
    dates = []
    feed = check_url(url_feed)
    for item in feed['items']:
        dates.append(item.get('published', ''))
    return dates

# Сбор данных до 1000 статей (с фильтром)
allheadlines = []
alldescriptions = []
alllinks = []
alldates = []
max_articles = 1000

for key, url in our_feeds.items():
    if len(allheadlines) >= max_articles:
        break
    headlines = getHeadlines(url)
    descriptions = getDescriptions(url)
    links = getLinks(url)
    dates = getDates(url)
    
    for h, d, l, dt in zip(headlines, descriptions, links, dates):
        if d is not None and len(allheadlines) < max_articles:  # Добавляем только с description
            allheadlines.append(h)
            alldescriptions.append(d)
            alllinks.append(l)
            alldates.append(dt)

# Сохранение в CSV
header = ['Title', 'Description', 'Link', 'Publication Date']
with open(f_all_news, 'w', encoding='utf-8-sig', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(header)
    for a, b, c, d in zip(allheadlines, alldescriptions, alllinks, alldates):
        writer.writerow([a, b, c, d])

print(f"Датасет сохранён в {f_all_news}. Собрано статей с description: {len(allheadlines)}")
df = pd.read_csv(f_all_news)  
```
